## **BackUp To Drive**

```python
from google.colab import drive
import shutil
import glob
import os


MODELNAME = MODELNAME  #### change
DRIVE_RVC_PATH = DRIVE_BACKUP_ROOT #### change
MODEL_DRIVE_FOLDER = f"{DRIVE_RVC_PATH}/{MODELNAME}"

os.makedirs(MODEL_DRIVE_FOLDER, exist_ok=True)
print(f"Backing up all files for {MODELNAME}...")

# 1. Backup Final Inference File (For the WebUI)
pth_source = f"/content/Retrieval-based-Voice-Conversion-WebUI/assets/weights/{MODELNAME}.pth"
if os.path.exists(pth_source):
    shutil.copy2(pth_source, MODEL_DRIVE_FOLDER)
    print("✅ Inference model weight backed up.")
else:
    print("⚠️ No Inference model weight backed up.")

# 2. Backup FAISS Index (For the WebUI)
index_files = glob.glob(f"/content/Retrieval-based-Voice-Conversion-WebUI/assets/indices/{MODELNAME}_IVF*.index")
if index_files:
    # Use follow_symlinks=True just in case, though we are targeting the real file now
    shutil.copy2(index_files[0], MODEL_DRIVE_FOLDER, follow_symlinks=True)
    print("✅ FAISS Index backed up.")
else:
    print("⚠️ No FAISS Index found to back up.")

# 3. Backup the LATEST Training Checkpoints (For Resuming)
# We only need the latest ones, not all of them!
log_dir = f"/content/Retrieval-based-Voice-Conversion-WebUI/logs/{MODELNAME}"

if os.path.exists(log_dir):
    # Find all G_*.pth and D_*.pth files
    g_files = glob.glob(f"{log_dir}/G_*.pth")
    d_files = glob.glob(f"{log_dir}/D_*.pth")

    if g_files and d_files:
        # Sort them numerically based on the epoch number to find the latest
        latest_g = sorted(g_files, key=lambda x: int(x.split('_')[-1].split('.')[0]))[-1]
        latest_d = sorted(d_files, key=lambda x: int(x.split('_')[-1].split('.')[0]))[-1]

        shutil.copy2(latest_g, MODEL_DRIVE_FOLDER)
        shutil.copy2(latest_d, MODEL_DRIVE_FOLDER)
        print(f"✅ Latest training checkpoints backed up: {os.path.basename(latest_g)} and {os.path.basename(latest_d)}")
    else:
         print("⚠️ No G/D checkpoints found to back up.")

    # Optional: Backup the config and filelist just in case
    if os.path.exists(f"{log_dir}/config.json"):
        shutil.copy2(f"{log_dir}/config.json", MODEL_DRIVE_FOLDER)
    if os.path.exists(f"{log_dir}/filelist.txt"):
        shutil.copy2(f"{log_dir}/filelist.txt", MODEL_DRIVE_FOLDER)

print(f"🎉 All backups complete! Stored in {MODEL_DRIVE_FOLDER}")
```

## **Restore From Drive For Interface**

```python
MODELNAME_TO_LOAD = MODELNAME #### change

print(f"Restoring {MODELNAME_TO_LOAD}...")

MODEL_DRIVE_FOLDER = f"{DRIVE_RVC_PATH}/{MODELNAME_TO_LOAD}"

if not os.path.exists(MODEL_DRIVE_FOLDER):
    print(f"❌ ERROR: Cannot find folder {MODEL_DRIVE_FOLDER} on Google Drive.")
else:
    # 1. Restore PyTorch weight into assets/weights
    pth_source = f"{MODEL_DRIVE_FOLDER}/{MODELNAME_TO_LOAD}.pth"
    pth_dest = f"/content/Retrieval-based-Voice-Conversion-WebUI/assets/weights/{MODELNAME_TO_LOAD}.pth"

    os.makedirs(os.path.dirname(pth_dest), exist_ok=True)
    if os.path.exists(pth_source):
        shutil.copy2(pth_source, pth_dest)
        print("✅ Restored PyTorch model weight.")
    else:
        print("⚠️ Warning: No .pth file found in Drive folder.")

    # 2. Restore FAISS index into assets/indices
    index_files = glob.glob(f"{MODEL_DRIVE_FOLDER}/{MODELNAME_TO_LOAD}_IVF*.index")
    index_dest_folder = "/content/Retrieval-based-Voice-Conversion-WebUI/assets/indices/"

    os.makedirs(index_dest_folder, exist_ok=True)
    if index_files:
        for index_file in index_files:
            shutil.copy2(index_file, index_dest_folder)
        print("✅ Restored FAISS index.")
    else:
        print("⚠️ Warning: No .index file found in Drive folder.")

    print("🎉 Restoration complete! You can now launch the UI.")
```
