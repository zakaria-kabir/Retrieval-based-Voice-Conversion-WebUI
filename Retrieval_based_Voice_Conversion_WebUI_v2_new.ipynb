```python
# @title ⚙️ RVC Training Settings

RVC_REPO = "/content/Retrieval-based-Voice-Conversion-WebUI"

MODELNAME = "sp001_test"  # @param {type:"string"}
DATASET_PATH = f"/content/dataset/{MODELNAME}"

SAMPLE_RATE = "48k"  # @param ["32k","40k","48k"]
SAMPLE_RATE_NUM = int(SAMPLE_RATE.replace("k", "")) * 1000

F0_METHOD = "rmvpe"  # @param ["rmvpe","harvest","dio"]

GPU = "0"  # @param {type:"string"}
BATCHSIZE = 2  # @param {type:"integer"}

TOTAL_EPOCH = 2  # @param {type:"integer"}
SAVE_EVERY = 1  # @param {type:"integer"}

DRIVE_BACKUP = "/home/zakaria/PersonalWork/Dialect_CLF_wav2vec2_v2/rvc_reference_audio/RVC_BackUPS"  # @param {type:"string"}

import os

# Get total CPU cores, subtract 1, ensure at least 1 thread
THREADS = max(1, os.cpu_count() - 1)
# THREADS = 8  # @param {type:"integer"}

import torch

DEVICE = f"cuda:{gpu_id}" if torch.cuda.is_available() else "cpu"
```

```python
# @title Preprocess dataset

! mkdir -p "$RVC_REPO/logs/$MODELNAME"
! [ -f "$RVC_REPO/logs/$MODELNAME/preprocess.log" ] || touch "$RVC_REPO/logs/$MODELNAME/preprocess.log"
! cd $RVC_REPO && python infer/modules/train/preprocess.py $DATASET_PATH $SAMPLE_RATE_NUM $THREADS $RVC_REPO/logs/$MODELNAME False 3.7
```

```python
# @title Extract f0
! cd $RVC_REPO && python infer/modules/train/extract/extract_f0_print.py $RVC_REPO/logs/$MODELNAME $THREADS $F0_METHOD


# @title Extract features
! cd $RVC_REPO && python infer/modules/train/extract_feature_print.py $DEVICE 1 0 0 $RVC_REPO/logs/$MODELNAME v2 True
```

**Train**

```python
# @title 1. First run the preparation script to generate filelist.txt and config.json
! cd $RVC_REPO && python prepare_train.py -e {MODELNAME} -sr {SAMPLE_RATE} -f0 1 -v v2
```

```python
import subprocess

result = subprocess.run(
    f"cd {RVC_REPO} && ls logs/{MODELNAME}/G_*.pth 2>/dev/null | sort -V | tail -n 1",
    shell=True,
    capture_output=True,
    text=True,
)
LATEST = result.stdout.strip()
if LATEST:
    print(f"Resuming from {LATEST}")
else:
    print("Starting new training")
```

```python
# @title 2. Train model

! cd $RVC_REPO && python infer/modules/train/train.py \
    -e {MODELNAME} \
    -sr {SAMPLE_RATE} \
    -f0 1 \
    -bs {BATCHSIZE} \
    -g {GPU} \
    -te {TOTAL_EPOCH} \
    -se {SAVE_EVERY} \
    -pg assets/pretrained_v2/f0G{SAMPLE_RATE}.pth \
    -pd assets/pretrained_v2/f0D{SAMPLE_RATE}.pth \
    -l 0 -c 0 -sw 1 -v v2
```
