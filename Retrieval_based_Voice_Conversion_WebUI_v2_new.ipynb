## **plot training metrics**

```python
import os, re
import pandas as pd
import matplotlib.pyplot as plt

logfile = "/content/Retrieval-based-Voice-Conversion-WebUI/logs/sp001/train.log"
print(os.path.exists(logfile))

rows = []

# Adjusted regex to match your log lines with timestamp and spaces
pattern = re.compile(
    r".*loss_disc=([\d\.]+),\s*loss_gen=([\d\.]+),\s*loss_fm=([\d\.]+),\s*loss_mel=([\d\.]+),\s*loss_kl=([\d\.]+)"
)

with open(logfile, "r", encoding="utf-8") as f:
    for ln in f:
        m = pattern.search(ln)
        if m:
            rows.append(
                {
                    "loss_disc": float(m.group(1)),
                    "loss_gen": float(m.group(2)),
                    "loss_fm": float(m.group(3)),
                    "loss_mel": float(m.group(4)),
                    "loss_kl": float(m.group(5)),
                }
            )

df = pd.DataFrame(rows)
if df.empty:
    print("No loss lines found. Check regex or log content!")
else:
    df.index.name = "step"  # just use row number as step
    df[["loss_mel", "loss_fm", "loss_gen", "loss_disc", "loss_kl"]].plot(
        subplots=False, figsize=(10, 5)
    )
    plt.yscale("linear")
    plt.grid(True)
    plt.show()
```

## **BackUp To Drive**

```python
from google.colab import drive
import shutil
import glob
import os


MODELNAME = MODELNAME  #### change
DRIVE_RVC_PATH = DRIVE_BACKUP_ROOT  #### change
MODEL_DRIVE_FOLDER = f"{DRIVE_RVC_PATH}/{MODELNAME}"

os.makedirs(MODEL_DRIVE_FOLDER, exist_ok=True)
print(f"Backing up all files for {MODELNAME}...")

# 1. Backup EVERY small inference weight (for testing quality)
# These are the small files (e100.pth, e120.pth) created by -sw 1
weight_files = glob.glob(
    f"/content/Retrieval-based-Voice-Conversion-WebUI/assets/weights/{MODELNAME}*.pth"
)
if weight_files:
    for f in weight_files:
        shutil.copy2(f, MODEL_DRIVE_FOLDER)
    print(
        f"✅ Backed up {len(weight_files)} inference weights (for testing different epochs)."
    )

# 2. Backup ONLY the LATEST giant training checkpoints (for resuming)
log_dir = f"/content/Retrieval-based-Voice-Conversion-WebUI/logs/{MODELNAME}"
if os.path.exists(log_dir):
    g_files = glob.glob(f"{log_dir}/G_*.pth")
    d_files = glob.glob(f"{log_dir}/D_*.pth")
    if g_files and d_files:
        latest_g = sorted(g_files, key=lambda x: int(x.split("_")[-1].split(".")[0]))[
            -1
        ]
        latest_d = sorted(d_files, key=lambda x: int(x.split("_")[-1].split(".")[0]))[
            -1
        ]
        shutil.copy2(latest_g, MODEL_DRIVE_FOLDER)
        shutil.copy2(latest_d, MODEL_DRIVE_FOLDER)
        print(f"✅ Latest resume-points backed up: {os.path.basename(latest_g)}")

# 3. Backup FAISS Index (Necessary for inference)
index_files = glob.glob(
    f"/content/Retrieval-based-Voice-Conversion-WebUI/logs/{MODELNAME}/added_IVF*.index"
)
if index_files:
    shutil.copy2(index_files[0], MODEL_DRIVE_FOLDER, follow_symlinks=True)
    print("✅ FAISS Index backed up.")

# 4. Backup Config/Filelist (Safety)
if os.path.exists(f"{log_dir}/config.json"):
    shutil.copy2(f"{log_dir}/config.json", MODEL_DRIVE_FOLDER)
    shutil.copy2(f"{log_dir}/filelist.txt", MODEL_DRIVE_FOLDER)

print(f"🎉 Backup Complete! All files are safe in: {MODEL_DRIVE_FOLDER}")
```

## **Restore From Drive For Interface**

```python
MODELNAME_TO_LOAD = MODELNAME  #### change

print(f"Restoring {MODELNAME_TO_LOAD}...")

MODEL_DRIVE_FOLDER = f"{DRIVE_RVC_PATH}/{MODELNAME_TO_LOAD}"

if not os.path.exists(MODEL_DRIVE_FOLDER):
    print(f"❌ ERROR: Cannot find folder {MODEL_DRIVE_FOLDER} on Google Drive.")
else:
    # 1. Restore PyTorch weight into assets/weights
    pth_source = f"{MODEL_DRIVE_FOLDER}/{MODELNAME_TO_LOAD}.pth"
    pth_dest = f"/content/Retrieval-based-Voice-Conversion-WebUI/assets/weights/{MODELNAME_TO_LOAD}.pth"

    os.makedirs(os.path.dirname(pth_dest), exist_ok=True)
    if os.path.exists(pth_source):
        shutil.copy2(pth_source, pth_dest)
        print("✅ Restored PyTorch model weight.")
    else:
        print("⚠️ Warning: No .pth file found in Drive folder.")

    # 2. Restore FAISS index into assets/indices
    index_files = glob.glob(f"{MODEL_DRIVE_FOLDER}/{MODELNAME_TO_LOAD}_IVF*.index")
    index_dest_folder = (
        "/content/Retrieval-based-Voice-Conversion-WebUI/assets/indices/"
    )

    os.makedirs(index_dest_folder, exist_ok=True)
    if index_files:
        for index_file in index_files:
            shutil.copy2(index_file, index_dest_folder)
        print("✅ Restored FAISS index.")
    else:
        print("⚠️ Warning: No .index file found in Drive folder.")

    print("🎉 Restoration complete! You can now launch the UI.")
```
